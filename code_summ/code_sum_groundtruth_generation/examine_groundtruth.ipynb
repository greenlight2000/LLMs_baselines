```python
import pandas as pd
gt = pd.read_csv("./task_selected_final_v1.csv")
gt
```

```python
err_indexs = gt.query("code_summarization=='Exception:Failed to generate'").index.tolist()
print(err_indexs)
```

```python
task_desc_len_list = []
for idx in err_indexs:
    task_desc = gt.loc[idx, "task_description"]
    task_desc_len_list.append(len(task_desc))
task_desc_len_df = pd.DataFrame(task_desc_len_list, columns=['len'])
task_desc_len_df.describe()
```

```python
task_desc_len_list.index(8002)
print(gt.loc[err_indexs[2], "task_description"])
```

```python
pass_task_desc_len_list = []
pass_indexs = gt.drop(err_indexs).index.tolist()
for idx in pass_indexs:
    task_desc = gt.loc[idx, "task_description"]
    pass_task_desc_len_list.append(len(task_desc))
pass_task_desc_len_df = pd.DataFrame(pass_task_desc_len_list, columns=['len'])
pass_task_desc_len_df.describe()
```

```python
i = pass_task_desc_len_list.index(499)
print(gt.loc[pass_indexs[i], "task_description"])
```
